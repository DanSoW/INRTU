```python
# Импорт модуля Faker
from faker import Faker
from faker.providers.company.ru_RU import Provider

# Импорт модуля random
import random
```

```python
# Импорт и настройка генератора случайных данных Faker
model = Faker('ru_RU')
model.add_provider(Provider)
```

```python
# Пустой список для хранения сгенерированных данных
fake_dataset1 = []

# Генерируем данные и добавляем их в список
for i in range(100):
  # Получаем рандомное число от 0 до 1
  k = random.randint(0, 1)

  if k == 1:
    # Если 1 - имя юридического лица
    name = model.company()
  else:
    # Если 0 - имя физического лица
    name = model.name()

  # Генерация банковского аккаунта
  bank_account = model.bban()

  # Добавление mock-данных в список
  fake_dataset1.append((i, name, bank_account))

# Создаём датафрейм вершин из списка данных
v = spark.createDataFrame(fake_dataset1, schema=['id', 'Client', 'Bank Account'])
```

```python
# Новый список для сгенерированных данных
fake_dataset2 = []

for i in range(200):
  # От кого
  src = random.randint(0, 19)
  # Кому
  dst = random.randint(0, 19)
  # Какая сумма
  summa = random.randint(0, 10000)

  fake_dataset2.append((src, dst, summa))

# Создаём датафрейм связей для списка данных
e = spark.createDataFrame(fake_dataset2, schema=['src', 'dst', 'summa'])
```

```python
e.count()

v.count()

v.coalesce(2).show()

e.coalesce(2).show()
```

```python
from pyspark import SparkConf, SparkContext

# Создание конфигурации
conf = SparkConf().set('spark.ui.port','4040').setAppName("Spark MapReduce").setMaster("local[2]")

# Создание новой Spark-сессии на основе конфигурации
spark = SparkContext.getOrCreate(conf=conf)
```

```python
# Создание 1 миллиона записей
fake = Faker('ru_RU')

# Записи (имена компаний)
data = []

for i in range(1000000):
  name = fake.company()
  data.append(name)
```

```python
# Создание распределённой базы данных
rdd = spark.parallelize(data, 4)
```

```python
# Отфильтруем данные по значению

rdd.filter(lambda company: company == 'Яндекс').collect()
```

```python
# Применим функцию map для данных с изменением регистра
rdd.map(lambda company: company.upper()).collect()
```

```python
# Сгруппируем значения по выбранному признаку
grouped_rdd = rdd.groupBy(lambda company: company[0]).map(lambda x: (x[0], list(x[1])))

grouped_rdd.collect()

rdd.groupBy(lambda company: company[0]).collect()
```

```python
from operator import add

b = rdd.map(lambda x: (x, 1)).reduceByKey(add).collect()

for (w, c) in b:
  print("{}: {}".format(w, c))
```
